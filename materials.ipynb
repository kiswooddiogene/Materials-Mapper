```python
import pandas as pd
from geopy import ArcGIS
import folium
nom = ArcGIS
```

1. **Import CSV**
2. **Display Entire Construction Materials Catalog**

```python
construction = pd.read_csv("construction-materials.csv")
print(construction.to_string())
```

**Enter all the materials you need seperated by comma, space**
1. *Each material should start with a CAPITAL Letter
2. *Ensure that each material is in the catalog
3. Don't forget to seperate the materials by spaces

```python
mats_list = []

needed_mats = input("Enter materials you need for your construction project: ")
seperate_mats = needed_mats.split(', ')

for items in seperate_mats:
        mats_list.append(items)
print(mats_list)

```

**Take your list of items you wanted and create a new DataFrame**

```python
mask = construction['Material'].isin(mats_list) 
mats_desired = construction[mask]
mats_desired
```

**Set starting location and create map** I will use NYC

```python
mats_map = folium.Map(location=[40.7128,-74.0060], zoom_start= 7)
mats_map
```

**Extract the Latitude and Longitude from the Table above using a for loop to mark each location and add parameters needed for map marker**

```python
from folium import plugins 
from folium.plugins import MiniMap

for (index, rows) in mats_desired.iterrows():
    lat = rows.loc['Latitude']
    long = rows.loc['Longitude']
    material = rows.loc['Material']
    country = rows.loc['Country']
    #folium.Marker(location=[lat,long],icon=folium.Icon(color='red'), popup=[f'Item: {material} \n Country: {country}.']).add_to(mats_map)
    
    radius = 20
    folium.CircleMarker(
    location=[lat,long],
    radius=radius,
    color="green",
    stroke=False,
    fill=True,
    fill_opacity=0.4,
    opacity=1,
    #popup=[f'Item: {material} \n Country: {country}.'].format(radius),
    tooltip=f'Item: {material} \n Country: {country}.]',
    ).add_to(mats_map)

MiniMap(toggle_display=True).add_to(mats_map)
    
```
